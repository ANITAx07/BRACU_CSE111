```python
#final summer
class Monster:
  monsterCount=0
  def __init__(self,name,power):
    self.name=name
    self.power=power
    self.alive=True
    Monster.monsterCount += 1
  def get_details(self):
    return f'Name:{self.name}\nPower:{self.power}\nAlive:{self.alive}'

  def attack(self,*args):
    if len(args) == 0 :
      print('No monsters to attack')
    for i in args:
      if self.alive == True and i.alive == True:
        if self.power>i.power:
          print(f'Attack successful.{self.name} defeated {i.name}.')
          i.alive = False
          Monster.monsterCount -= 1
        else:
          print(f'{self.name} defeated {i.name}')
          Monster.monsterCount -= 1
      elif self.alive == True and i.alive == False:
        print(f"Cannot attack {i.name} \nIt's not alive.\nAttack unsuccessful.")
      elif self.alive == False:
        print(f'{self.name} is not alive to attack.')


monster1 = Monster('Godzilla', 40)
monster2 = Monster('Hydra', 30)
monster3 = Monster('KingKong', 50)
print(f"Number of monsters alive:{Monster.monsterCount}")
print('1--------------------------------')
print(monster1.get_details())
print('2--------------------------------')
print(monster2.get_details())
print('3--------------------------------')
print(monster3.get_details())
print('4--------------------------------')
monster1.attack()
print('5--------------------------------')
monster1.attack(monster2)
print('6--------------------------------')
monster1.attack(monster2, monster3)
print('7--------------------------------')
print(f"Number of monsters alive:{Monster.monsterCount}")
print('8--------------------------------')
print(monster2.get_details())
print('9--------------------------------')
monster2.attack(monster1)
```
